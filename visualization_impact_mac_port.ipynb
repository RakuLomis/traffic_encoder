```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
from evaluation.confusion_matrix_visualization import plot_confusion_matrix_panel_gamma, plot_confusion_matrix_panel
```

```python
cm_dir = os.path.join('..', 'evaluation', 'confusion_matrix') 
prefix = '0.5s_flow_noip_' 
postfix = '.csv' 
output_pdf_path = os.path.join(cm_dir, 'impact_mac_cm_.pdf')
abalation_name_list = ['eth_port', 'eth_noport', 'noeth_port', 'noeth_noport'] 
label_order = ['Youtube','Twitter','Spotify','Facebook','Vimeo','Google']
```

```python
csv_paths = []
for e_name in abalation_name_list: 
    csv_path = os.path.join(cm_dir, prefix + e_name + postfix)
    csv_paths.append(csv_path)

cm_list = [pd.read_csv(path, index_col=0) for path in csv_paths]


cm_list = [
    cm.loc[label_order, label_order]
    for cm in cm_list
]

titles = [
    "ETH & Port",
    "ETH only",
    "Port only",
    "Neither"
]
```

```python

fig = plot_confusion_matrix_panel(
    cm_list=cm_list,
    titles=titles,
    label_order=label_order,
    cell_scale=0.6  # 🔥 更小
)
plt.savefig(output_pdf_path, format="pdf", bbox_inches='tight')
print("PDF is saved.")
plt.show()
plt.close()
```
