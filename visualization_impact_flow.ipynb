```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
from evaluation.confusion_matrix_visualization import plot_confusion_matrix_panel_gamma, plot_confusion_matrix_panel
```

```python
cm_dir = os.path.join('..', 'evaluation', 'confusion_matrix') 
# prefix = '0.5s_flow_noip_' 
# prefixs = ['0.05s_noflow_', '0.05s_flow_', '0.1s_noflow_', '0.1s_flow_', '0.5s_noflow_', '0.5s_flow_']
postfix = 'noip_eth_port.csv' 
output_pdf_path1 = os.path.join(cm_dir, 'impact_flow_row1.pdf')
output_pdf_path2 = os.path.join(cm_dir, 'impact_flow_row2.pdf')
abalation_name_list_r1 = ['0.05s_noflow_', '0.1s_noflow_', '0.5s_noflow_'] 
abalation_name_list_r2 = ['0.05s_flow_',  '0.1s_flow_', '0.5s_flow_'] 
label_order = ['Youtube','Twitter','Spotify','Facebook','Vimeo','Google']
```

```python
csv_paths = []
for e_name in abalation_name_list_r1: 
    csv_path = os.path.join(cm_dir, e_name + postfix)
    csv_paths.append(csv_path)

cm_list = [pd.read_csv(path, index_col=0) for path in csv_paths]


cm_list = [
    cm.loc[label_order, label_order]
    for cm in cm_list
]

titles = [
    "5% Sampling",
    "10% Sampling",
    "50% Sampling",
]
```

```python

fig = plot_confusion_matrix_panel(
    cm_list=cm_list,
    titles=titles,
    label_order=label_order,
    cell_scale=0.6  # 🔥 更小
)
plt.savefig(output_pdf_path1, format="pdf", bbox_inches='tight')
print("PDF is saved.")
plt.show()
plt.close()
```

```python
csv_paths = []
for e_name in abalation_name_list_r2: 
    csv_path = os.path.join(cm_dir, e_name + postfix)
    csv_paths.append(csv_path)

cm_list = [pd.read_csv(path, index_col=0) for path in csv_paths]


cm_list = [
    cm.loc[label_order, label_order]
    for cm in cm_list
]

titles = [
    "5% Sampling & Flow Expert",
    "10% Sampling & Flow Expert",
    "50% Sampling & Flow Expert"
]
```

```python
fig = plot_confusion_matrix_panel(
    cm_list=cm_list,
    titles=titles,
    label_order=label_order,
    cell_scale=0.6  # 🔥 更小
)
plt.savefig(output_pdf_path2, format="pdf", bbox_inches='tight')
print("PDF is saved.")
plt.show()
plt.close()
```
