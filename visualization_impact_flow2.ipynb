```python
csv_paths = []
for e_name in abalation_name_list: 
    csv_path = os.path.join(cm_dir, datasets[0] + '_' +  e_name + postfix)
    csv_paths.append(csv_path)

cm_list = [pd.read_csv(path, index_col=0) for path in csv_paths]


cm_list = [
    cm.loc[label_order_CSTNET, label_order_CSTNET]
    for cm in cm_list
]

titles = [
    "w SII",
    "w/o SII",
]
```

```python
fig = plot_confusion_matrix_global_representative(
    cm_list=cm_list,   # DataFrame list
    titles=titles,
    repr_labels=cstnet_repr_labels,
    tick_fontsize=12,
    title_fontsize=14,
    cell_scale=0.28,
    global_xlabel_y=0.08, 
    annot=True,
    annot_fontsize=7,
    use_fixed_figsize=True, 
    fixed_figsize=(8.0, 4.2), 
    matrix_only=True
)
plt.savefig(os.path.join(cm_dir, 'impact_flow_cstnet_cm.pdf'), format="pdf", bbox_inches='tight')
print("PDF is saved.")
plt.show()
plt.close()
```
